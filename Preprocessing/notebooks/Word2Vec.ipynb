```python
#import
import numpy as np
import nltk
import string
from nltk.tokenize import TweetTokenizer
from collections import Counter
import pandas as pd
import gensim.models
import pickle
import csv
```

### Load Datasets

```python
pos = [line.rstrip('\n') for line in open('../../data/filtered_data/cleaned_train_pos.txt')]
neg = [line.rstrip('\n') for line in open('../../data/filtered_data/cleaned_train_neg.txt')]
test = [line.rstrip('\n') for line in open('../../data/filtered_data/filtered_test_data.txt')]

pos[:10], neg[:10], test[:10]
```

```python
test = [line.rstrip('\n') for line in open('../../data/filtered_data/filtered_test_data.txt')]
```

```python
test = [tweet if tweet != '' else 'empty' for tweet in test]
```

```python
pos = [tweet if tweet != '' else 'empty' for tweet in pos]
neg = [tweet if tweet != '' else 'empty' for tweet in neg]
test = [tweet if tweet != '' else 'empty' for tweet in test]
```

```python
tknzr = TweetTokenizer()
pos_tokens = list([tknzr.tokenize(tweet) for tweet in pos])
neg_tokens = list([tknzr.tokenize(tweet) for tweet in neg])
test_tokens = list([tknzr.tokenize(tweet) for tweet in test])
```

```python
# Using the full dataset
pos_full = [line.rstrip('\n') for line in open('../../data/filtered_data/cleaned_train_pos_full.txt')]
neg_full = [line.rstrip('\n') for line in open('../../data/filtered_data/cleaned_train_neg_full.txt')]

pos_full = [tweet if tweet != '' else 'empty' for tweet in pos_full]
neg_full = [tweet if tweet != '' else 'empty' for tweet in neg_full]

pos_full_tokens = list([tknzr.tokenize(tweet) for tweet in pos_full])
neg_full_tokens = list([tknzr.tokenize(tweet) for tweet in neg_full])
```

# Word Embeddings

### Genrate word2vec model

```python
test_tokens
```

```python
model = gensim.models.Word2Vec(test_tokens, size=200, window=5, min_count=1, workers=4)
```

```python
model.wv.similarity('sad', 'night')
```

```python
model.wv.vocab
```

```python
# Using the full dataset
model_full = gensim.models.Word2Vec(pos_full_tokens + neg_full_tokens + test_tokens, size=100, window=5, min_count=1, workers=4)
```

# Tweets to vectors

```python
def tweet2vector(tweet, model):
    return sum([model[word] for word in tknzr.tokenize(tweet)])
```

```python
pos_vec = np.asarray([tweet2vector(tweet, model) for tweet in pos])
neg_vec = np.asarray([tweet2vector(tweet, model) for tweet in neg])

# Concatenate both
X = np.vstack((pos_vec, neg_vec))
```

```python
len(X)
```

```python
#labels
y = np.concatenate((np.ones(len(pos_vec)), -1*np.ones(len(neg_vec))))
```

```python
# Testing set
test_vec = np.asarray([tweet2vector(tweet, model) for tweet in test])
```

```python
# Using the full dataset
pos_full_vec = np.asarray([tweet2vector(tweet, model_full) for tweet in pos_full])
neg_full_vec = np.asarray([tweet2vector(tweet, model_full) for tweet in neg_full])

# Concatenate both
X_full = np.vstack((pos_full_vec, neg_full_vec))

y_full = np.concatenate((np.ones(len(pos_full_vec)), -1*np.ones(len(neg_full_vec))))
```

```python
def create_csv_submission(ids, y_pred, name):
    """
    Creates an output file in csv format for submission to kaggle
    Arguments: ids (event ids associated with each prediction)
               y_pred (predicted class labels)
               name (string name of .csv output file to be created)
    """
    idx = pickle.load( open( "../../data/submissions/idx.p", "rb" ))+1
    name += str(idx)
    with open(name, 'w') as csvfile:
        fieldnames = ['Id', 'Prediction']
        writer = csv.DictWriter(csvfile, delimiter=",", fieldnames=fieldnames)
        writer.writeheader()
        for r1, r2 in zip(ids, y_pred):
            writer.writerow({'Id':int(r1),'Prediction':int(r2)})
    pickle.dump( idx, open( "../../data/submissions/idx.p", "wb" ))
```

```python
create_csv_submission(np.arange(len(predictions))+1, predictions, '../../data/submissions/prediction')
```
