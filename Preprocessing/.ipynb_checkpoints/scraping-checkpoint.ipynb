```python
# Import libraries
import requests
from bs4 import BeautifulSoup
```

```python
def write_to_file(word_list, file):
    path = '../../data/'+file
    f = open(path, 'w', encoding='utf-8') 
    for word in word_list:
        f.write(word + '\n')
    f.close()
    print('the file has been successfully created in :: ', path)
```

```python
def scrap_url(URL):
    r = requests.get(URL)
    page_body = r.text
    soup = BeautifulSoup(page_body, 'html.parser')
    li = soup.find_all('li')
    symbols = []
    for p in li:
        s = p.find('span')
        if s != None:
            text = p.getText(separator=u'_______').replace('\t','').lower().split('_______')
            symbols.append(text[0].replace(' ',''))
            symbols.append(text[1])
    return symbols
```

```python
URL = 'http://www.netlingo.com/acronyms.php'
acronyms = scrap_url(URL)
acronyms[:10]
```

```python
write_to_file(acronyms, 'netlingo_acronyms.txt')
```

```python
URL = 'http://www.netlingo.com/smileys.php'
smileys = scrap_url(URL)
smileys[:10]
```

```python
write_to_file(smileys, 'netlingo_smileys.txt')
```

```python
URL = 'https://en.wiktionary.org/wiki/Wiktionary:Frequency_lists/PG/2006/04/30001-40000'
```

```python
def scrap_dict(URL):
    r = requests.get(URL)
    page_body = r.text
    soup = BeautifulSoup(page_body, 'html.parser')
    tr = soup.find_all('tr')
    words = []
    frequencies = []
    text_list = []
    for p in tr[1:]:
        s = p.find('td')
        if s != None:
            text = p.getText(separator=u'_______').replace('\t','').lower().split('_______')
            text = [t for t in text if t != '\n']
            text_list.append(text)
            words.append(text[1].replace(' ',''))
            frequencies.append(text[2])
            
    return dict(zip(words, frequencies)), text_list
```

```python
URL = 'https://en.wiktionary.org/wiki/Wiktionary:Frequency_lists/PG/2006/04/30001-40000'
english_dict = []
text_list = []
for i in range(4):
    URL = 'https://en.wiktionary.org/wiki/Wiktionary:Frequency_lists/PG/2006/04/{}-{}'.format(10000*i+1, 10000*(i+1))
    english_dict.append(scrap_dict(URL)[0])
    text_list.append(scrap_dict(URL)[1])
```
